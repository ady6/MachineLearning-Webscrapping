#### En un DataLake (explica qué es) selecciona una fuente de datos. Dataset o similar, en csv, json…

Un DataLake es un repositorio de almacenamiento que contienen una gran cantidad de datos en bruto y que se mantienen allí hasta que sea necesario.

Como fuente de datos he utilizado un Dataset en json llamado 'Alumnos.json' que creé yo misma con Java y subí a mi repositorio de GitHub: https://github.com/ady6/MachineLearning-Webscrapping/blob/82fb8b5ebc71e8efd6c3c592ae04af740a122cbc/Alumnos.json.

Este contiene un total de 6 columnas (id_Alumno, nombre, apellidos1, apellidos2, grupo, fecha_nacimiento) las cuales son el id, nombre, apellido paterno, apellido materno, grupo de clase y fecha de nacimiento del alumno. También tiene un total de 101 filas ya que los alumnos van de 0 a 100.

#### Utilizando Jupyter crea un notebook que permita acceder a la fuente de datos anterior y comprueba que estás mostrando los 10 primeros registros.

```python
#importo la libreria pandas para manejar datos
import pandas as pd #con el alias pd

#ruta de archivo json, el cual se encuentra en mi repositorio de GitHub
url="https://raw.githubusercontent.com/ady6/MachineLearning-Webscrapping/master/Alumnos.json"
df = pd.read_json(url) #leo el archivo con pandas (pasando su url)
#muestro los datos del archivo realizando su debida conversión de tipos de datos para su correcto funcionamiento (to_string)
print(df.to_string())
```

```python
#utilizo el método head() de Pandas para mostrar las primeras n filas (en este caso 10) del dataframe 
df.head(10)
```

#### Utiliza el método describe e info de Python y explica brevemente los resultados.

```python
df.describe()
```

El método describe devuelve información estadística de los datos del dataframe (de hecho, este método devuelve un dataframe). En este caso, esta información incluye el número de alumnos (101 ya que van de 0 a 100), el valor medio (en este caso 50), la desviación estándar (indica cuánto se agrupan los datos alrededor de la media de estos), el valor mínimo (en este caso el alumno de id 0), máximo (el alumno con id 100) y los valores correspondientes a los percentiles 25%, 50% y 75% (en este dataframe conocer estos valores es muy poco útil ya que serán idénticos a los porcentajes ya que se hacen sobre 100 alumnos y el 100% es el porcentaje total).

```python
df.info()
```

El método info muestra el resumen del dataframe.
Como podemos ver, devuelve la siguiente información: número de filas (de 0 a 100, en total 101 filas), número y nombre de columnas (en este caso 6:  id_Alumno, nombre, apellidos1, apellidos2, grupo, fecha_nacimiento), índices de las filas, valores no nulos, tipo de las columnas (el id como entero y el resto como objetos) y su cantidad (1 int y 5 object), y la memoria usada (4.9KB).
