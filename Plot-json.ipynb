Ejercicio2
conéctate a un json y muestra algunos datos en python para probar que funciona

```python
#importo las librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt

url="https://www.w3schools.com/python/pandas/data.js" #ruta de archivo json
df = pd.read_json(url) #leo el archivo con pandas (pasando la url)
print(df.to_string()) #muestro los datos del archivo
```

Ejercicio3
representa algunos datos en un gráfico circular

```python
df.Maxpulse.value_counts().plot.pie() #cuento los valores del pulso máximo de cada registro
plt.show() #los muestro
```
