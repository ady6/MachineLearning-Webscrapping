```python
#importo lo necesario
import requests
from bs4 import BeautifulSoup
#establezco la url de la página web a la que realizar la petición
ruta="http://www.bne.es/es/Inicio/index.html" #es la web de la Biblioteca Nacional de España
#realizo una petición get a la web
contenido=requests.get(ruta)
#muestro el contenido de la web en formato html
soup = BeautifulSoup(contenido.content,"html.parser")
print(soup)
```

```python
#muestro enlaces que contiene la web
#para ello accedo a la etiqueta a y muestro su atributo href
print(soup.a['href'])
```

```python
#muestro el contenido de las etiquetas HTML con un id header
cabecera=soup.find(id="header")
print(cabecera)
```
