```python
#importo sqlite3 para realizar la conexion con la bbdd
import sqlite3
#creo la conexion con la bbdd
conexion=sqlite3.connect('sample.db')
#creo el cursor
cursor=conexion.cursor()
#realizo un select para mostrar los datos de la tabla customers
consulta="select * from customers"
#ejecuto la consulta con el cursor
cursor.execute(consulta)
#obtengo los resultados
resultado=cursor.fetchall()
#los muestro
print(resultado)
```

```python
#importo pandas con alias pd para crear un dataframe
import pandas as pd
cursor=conexion.cursor()
#consulta para mostrar el nombre, codigo postal, ciudad y pais en orden descendente por ciudad
consulta="select FirstName,PostalCode,City,Country from customers order by City desc"
cursor.execute(consulta)
resultado=cursor.fetchall()
#en un dataframe almaceno los datos del resultado
df=pd.DataFrame(resultado)
#lo muestro
print(df)
```

```python
#importo el modulo pyplot de la libreria matplotplib para hacer el grafico circular
import matplotlib.pyplot as plt
#consulta para mostrar numero de clientes por pais (unicamente los 5 primeros paises con mas clientes)
consulta="select Country, count(CustomerId) as totalClientes from customers group by Country order by totalClientes desc limit 5"
cursor.execute(consulta)
resultado=cursor.fetchall()
#en un dataframe almaceno los datos del resultado
df=pd.DataFrame(resultado)
#creo el grafico circular
plt.pie(df[1],labels=df[0])
#muestro el grafico
plt.show()
```

```python
#consulta para mostrar numero de clientes por pais
consulta="select Country, count(CustomerId) as totalClientes from customers group by Country"
cursor.execute(consulta)
resultado=cursor.fetchall()
#en un dataframe almaceno los datos del resultado
df=pd.DataFrame(resultado)
#creo el grafico de puntos
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x = df[1], y = df[0])
#muestro el grafico
plt.show()
```
