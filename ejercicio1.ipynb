```python
#importo la librería pandas
import pandas as pd

#csv descargado de: https://www.kaggle.com/datasnaek/chess
#leo el csv con pandas
df=pd.read_csv(r'https://raw.githubusercontent.com/ady6/MachineLearning-Webscrapping/master/games.csv')

#muestro el archivo
print(df)
```

```python
#muestro el archivo en gráfica
df.plot(kind ="hist")
```

```python
#muestro clasificacion de piezas blancas y piezas negras
import matplotlib.pyplot as plt
my_plot = df.plot("black_rating", "white_rating", kind="scatter")
plt.show()
```

```python
#gráfica frecuencia de turnos
df.turns.plot.hist()
plt.show()
```
